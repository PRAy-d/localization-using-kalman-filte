## Update

```python
# defining a function update, that updates the value of mesurements with 
# the covariance and mean of previous measuremennts

# variables
#  u1: mean of original measurememnt
#  var1: variance of original measurement
#  u1: mean of new measurements
#  var2: variance of new measurements
#  new_mean: updated mean of measurements
#  new_var: updated variance of measurements

# ouput of function
#  updated values of mean and variance of measurements


def update(u1,var1,u2,var2):
    new_mean=(var2*u1+var1*u2)/(var1+var2)
    new_var=1/(1/var1+1/var2)
    return[new_mean,new_var]
```

## Predict

```python
# defining a function predict, that predicts the value of the motion with 
# the covariance and mean values of previous motion

# variables
#  u1: mean of original motion values
#  var1: variance of original motion values
#  u1: mean of new motion value
#  var2: variance of new motion value
#  new_mean: updated mean of motion values
#  new_var: updated variance of motion values

# ouput of function
#  updated values of mean and variance of mootion

def predict(u1,var1,u2,var2):
    new_mean=u1+u2
    new_var=var1+var2
    return [new_mean,new_var]
```

## 1D kalman filter

```python
# measurements[n] is the output values from the sensors
# motion[n] is the location value of bot
#measurements=[]
#motion=[]
u=0
sig=100
meas_var=100
mot_var=100
n=len(measurements)
for i in range(n):
    [u,sig]=update(u,sig,measurements[i],meas_var)
    print('update: ',[u,sig])
    [u,sig]=predict(u,sig,motion[i],mot_var)
    print('update: ',[u,sig])
```
